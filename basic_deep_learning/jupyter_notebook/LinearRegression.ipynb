```python
# Linear Regression

import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt

# ta dùng công thức MSE để tính loss function

data = pd.read_csv("../data_linear.csv")
# print(data.values)
# data ở đây là 1 class Object, không phải là list
# data.values sẽ return lại 1 Array gôm các phần tử trong đó theo dạng [][]
# dữ liệu diện tích 
x = data.values[:,0].reshape((-1,1))
# dữ liệu giá nhà thực tế
y = data.values[:,1].reshape((-1,1))

N = data.shape[0]
```

```python
# Công thức MSE 
# L = MSE = sigma(i=1,n) ((Yi - Yip)^2) / N. Trong đó Yip = w0 + w1.Xi
# Yi và Xi là hằng số thực tế có sẵn
# w0 và w1 là ẩn số mà bài toán phải tìm ra
```

```python
learning_rate = 0.00001
iteration = 100

x = np.hstack((np.ones((N,1)),x))
w = np.array([[0],[1]]).astype('float64')
for i in range(1, iteration):
    # Y predict
    yP = np.dot(x, w)
    r = y - yP
    w[0] += learning_rate * np.sum(r)
    w[1] += learning_rate * np.sum(np.multiply(x[:,1].reshape((-1,1)),r))
print(w)
# Khi ẩn số w ko thể thay đổi được nữa thì ta tiến hành bước dự đoán từ dữ liệu thực tế
predict = np.dot(x,w)
plt.scatter(x[:,1], y)
plt.xlabel('Meter')
plt.ylabel('Price')
plt.plot((x[0][1], x[N-1][1]), (predict[0], predict[N-1]),'r')
plt.show()
```
